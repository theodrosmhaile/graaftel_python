```python
import graaftel as gt
import numpy as np
import pandas as pd
import math
import random
```

### Train a new model

```python
m = gt.Model() # make model object

# add in data file with 3 columns: 
#  student Id | Question Id | Score 
# Note that column headers should not be included

m.init_model(data='test_data/small_dataset_for_testing.csv', nEpochs=1000) ## at least about a 1000 epochs is safe 

m.get_ratings()

## can save model here if needed. 
#m.save(file_name='test_data_model_4skills')
```

```python
### to see a small graph of the trained data(library networkx is required for plotting to work)
gt.make_nodes(m, graph=True)
```

### Run an individual student

```python
#set student mode to True
m.studentMode = True
## If a model is being imported/loaded studentMode will be set at import using m.load_model()
```

```python
# if the student does not exist:
m.new_student(name='Teddy')
print('New student skills:', m.students['Teddy'].skills)
```

```python
# update student score
m.update_student_rating(name='Teddy', item='10', score=1)
print('Updated student skills:', m.students['Teddy'].skills)
```
